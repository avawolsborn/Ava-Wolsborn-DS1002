1.  Load the following dataset into Colab and perform the requested operations in code.

 Source: https://ds1002-resources.s3.amazonaws.com/data/ride_sharing.csv

```python
# Load pandas and the data

import pandas as pd
ride_shares = pd.read_csv('ride_sharing.csv')
```

```python
# show how many rows and columns are in this dataset

ride_shares.shape

# there are 25760 rows and 10 columns.
```

```python
# displaying the first few rows to visualize data:
ride_shares.head(10)
```

```python
# The first column, "duration" is seen as a string.
# Using two snippets of code, clean this column up so that
# is now an integer (int64)

ride_shares['duration'] = ride_shares['duration'].str.strip(' minutes')
ride_shares['duration'] = ride_shares['duration'].astype(int)
```

```python
# Now display the datatypes in the updated dataframe

ride_shares.dtypes
```

```python
# Show the min and max values of ride durations (the shortest and longest rides)
# Hint: If 'df' is the name of your dataframe, simply specify the column name
# and then append the .min() or .max() methods.

print("The minimum duration of ride shares is " + str(ride_shares.duration.min()) + " minute")
print("The maximum duration of ride shares is " + str(ride_shares.duration.max()) + " minutes")
```

```python
# Create a new dataframe object and populate it with rides that are 60 minutes or less.
# Hint: Use the .query() method for the dataframe that we used in class.

short_rides = ride_shares.query('duration <= 60')
short_rides
```

```python
# Now show both the min and max values of the duration column for your new dataset.

print("The minimum duration of ride shares is " + str(short_rides.duration.min()) + " minute")
print("The maximum duration of ride shares is " + str(short_rides.duration.max()) + " minutes")
```

```python
# Using the new dataframe, update it so that is ONLY has records where
# the tire_size is 26.

tire_26 = short_rides.query('tire_size == 26')
tire_26
```

```python
# Finally, produce a count of all records with FEMALE riders
# using the 26-inch wheeled rental bikes for their rentals of
# one hour or less.

female_riders = tire_26.query('user_gender == "Female"')
female_riders

#There are 2962 rows in this dataframe, indicating the count of all female riders
# using 26-inch wheeled rental bikes for rentals of an hour or less
```

2. Find the file `more-messy-data.csv` on GitHub and upload it into Google Colab.

Create a new Pandas DataFrame from that data and perform the following steps.

```python
# Write a snippet of code that will display the number of duplicate rows exist in the DF.

data = pd.read_csv('more-messy-data.csv')
print(data)

print(data.duplicated().sum())
```

```python
# Next, remove all duplicate rows using Python.

data = data.drop_duplicates()
data
```

```python
# For the FOUR columns involving sepal and petal length and width, update all NaN
# values with the column mean.

data = data.fillna(data.mean())
data
```

```python
# Find any values in the dataframe that are still missing, and delete them.

data = data.dropna()
data
```

```python
# How many different species are there?

print(data['species'].describe())

# There are 9 different species, but the data has yet to be cleaned, so we cannot guarantee this is true
```

```python
### Clean up the 'species' variable so that there are exactly three different values.

# ensures that all species have the same capitalization
data["species"] = data["species"].str.lower()

# counting how many unique species there are after this cleaning functino
print(data['species'].value_counts())

# assigning the species into categories, considering certain species had only 1 or 2 (likely mispellings)
unique_species = {'setosa': 'setosa', 'virginica': 'virginica', 'versicolor': 'versicolor', 'setoso': 'setosa', 'virginia': 'virginica', 'viginica': 'virginica', 'setosan': 'setosa'}

# use `.replace` to filter into categories
data['species'] = data['species'].replace(unique_species)
print(data['species'].unique())
print(data['species'].describe())

print(data)
```

The `re` package has various functions related to regular expressions, including for example `match()`, which takes two arguments, first the regular expression and second the test string, and tests to see whether the *beginning* of the `test` string matches the pattern `exp`.

Note that `re.match(exp,test)` does not return a boolean variable, as you might expect (in other words, a match doesn't return `True` and a non-match `False`).  The type of object you get is called `re.Match`, which is a data type special to the `re` package.
