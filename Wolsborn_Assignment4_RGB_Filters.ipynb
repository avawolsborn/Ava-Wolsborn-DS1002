```python
import numpy as np
from PIL import Image
import requests
import matplotlib.pyplot as plt
from skimage import io
```

```python
# prompt the user for the url of an image
print("Insert the url to a .jpg file:")
url = input()

# cast the url as a numpy array, define as a variable 'RGBimage'
RGBimage = np.asarray(Image.open(requests.get(url, stream=True).raw))

## function to filter the image red
def redfilter():
  # ensure that the variable 'photo_red' is defined outside of this function by assigning it as global
  global photo_red
  # create a photo variable that reads the url, then copy it so we don't edit the original
  photo = io.imread(url)
  photo_red = photo.copy()
  # red, green, blue have indices of 0, 1, 2 -- set the appropriate indices equal to 0
  photo_red[:,:,1] = 0    #sets green to 0
  photo_red[:,:,2] = 0    #sets blue to 0
  # return the global variable for printing at the end
  return photo_red

### repeat the above process with all color combinations

def greenfilter():
  global photo_green
  photo = io.imread(url)
  photo_green = photo.copy()
  photo_green[:,:,0] = 0    #sets red to 0
  photo_green[:,:,2] = 0    #sets blue to 0
  return photo_green

def bluefilter():
  global photo_blue
  photo = io.imread(url)
  photo_blue = photo.copy()
  photo_blue[:,:,0] = 0    #sets red to 0
  photo_blue[:,:,1] = 0    #sets green to 0
  return photo_blue

def bluefilter():
  global photo_blue
  photo = io.imread(url)
  photo_blue = photo.copy()
  photo_blue[:,:,0] = 0    #sets red to 0
  photo_blue[:,:,1] = 0    #sets green to 0
  return photo_blue

def redgreenfilter():
  global photo_RG
  photo = io.imread(url)
  photo_RG = photo.copy()
  photo_RG[:,:,2] = 0    #sets blue to 0
  return photo_RG

def redbluefilter():
  global photo_RB
  photo = io.imread(url)
  photo_RB = photo.copy()
  photo_RB[:,:,1] = 0    #sets green to 0
  return photo_RB

def greenbluefilter():
  global photo_GB
  photo = io.imread(url)
  photo_GB = photo.copy()
  photo_GB[:,:,0] = 0    #sets red to 0
  return photo_GB

# call each function in order to filter and retrieve the photos
redfilter()
greenfilter()
bluefilter()
redgreenfilter()
redbluefilter()
greenbluefilter()

# create an array with every filtered image
# this ensures that every photo will print, not just the last one
images = [RGBimage, photo_red, photo_green, photo_blue, photo_RG, photo_RB, photo_GB]

# for loop will cycle through every item in the above list and print it
for ima in images:
    plt.figure()
    plt.imshow(ima)
```

The below code is ensuring that the concatenation of the red, blue, and green image arrays are identical to the original image's array

```python
RGBimage == photo_red + photo_green + photo_blue
```
