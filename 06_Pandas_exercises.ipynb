10.  Load the `iris` dataset that we used in class.  Use `iris` as the name of your data frame.

```python
import seaborn as sns
iris = sns.load_dataset('iris')
```

11.  What is the correlation between the variables `sepal_length` and `sepal_width` for all the irises?

```python
iris[['sepal_length','sepal_width']].corr()
```

There is a correlation of -0.11757 between the variables sepal_length and sepal_width.

12.  Which of the three species has the highest correlation between `sepal_length` and `sepal_width`?  Show any calculations you use to support your answer.

```python
## For each species:
# define a species-specific data frame that prints the information from iris of only that species
# print the species name and the correlation values of the two variables for that specific species: sepal_length and sepal_width

setosa_species = iris[iris.species == 'setosa']
print("Setosa:")
print(setosa_species[['sepal_length','sepal_width']].corr())

versicolor_species = iris[iris.species == 'versicolor']
print("Versicolor:")
print(versicolor_species[['sepal_length','sepal_width']].corr())

virginica_species = iris[iris.species =='virginica']
print("Virginica:")
print(virginica_species[['sepal_length','sepal_width']].corr())
```

The setosa species has the highest correlation between the variables sepal_length and sepal_width.

Determine the same thing but for each species separately:  for each species, which two variables have the most negative (or the least positive) correlation? 14.  Once again, the variables `sepal_width` and `petal_length` are negatively correlated.  But for each individual species, these two variables are positively correlated.  How is this possible?  Explain this phenomenon.  Use a data visualization if it helps.


This is possible because with each individual species, as the sepal length increases, the sepal width generally does as well. However, every species holds a different average sepal length and width. For example, the sepal length of virginicas is greater than that of setosas, but the average sepal width is less than that of setosas. As there is an increase in the average sepal length between species, there is a decrease in the average sepal width for each.

With that, we find a data visualization that will appear to be increasing in 3 separate sections but decreasing in values when looking across all 3 species. As a result, we end up with an overall negative correlation and a positive correlation within each species.
